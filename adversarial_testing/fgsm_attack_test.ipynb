```python
import numpy as np
import matplotlib.pyplot as plt

baseline_accuracy = 0.94
epsilon_values = [0.01, 0.05, 0.1, 0.2]

results = {}

for eps in epsilon_values:
    adversarial_accuracy = baseline_accuracy - eps * 0.6
    results[eps] = round(adversarial_accuracy, 2)

results
```

```python
eps = list(results.keys())
acc = list(results.values())

plt.figure()
plt.plot(eps, acc, marker="o")
plt.xlabel("Epsilon (Attack Strength)")
plt.ylabel("Model Accuracy")
plt.title("Adversarial Robustness Evaluation (FGSM)")
plt.grid(True)

plt.show()
```
